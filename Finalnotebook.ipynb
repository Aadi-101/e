# Clustering Neighbourhoods to Suitable Locations for new bookstores in Mumbai

# Introduction

In the densely populated and highly competitive city of Bombay, business wise, if new businesses or franchises have to be established, looking for suitable locations becomes very important, if i.e. the owner wants his business to thrive.

When it comes to bookstores, no ine visits them as frequently as school children and college students.
Furthurmore, genrally avid readers, barring students, in such a economically diverse city such as Bombay, placing bookstores within the vicinity of premium coffee shops can be a fairly advantageous to both coffee and book store owners as if a fairly affluent individual goes out for either one of them, he/she is more likely to go for the other, as reading and drinking coffee is one of the most popular passtimes within the upper middle class and overtly affluent demographics.
Bombay being the economic hub of India, there is absolutely no shortage of premium coffee joints, and as a matter of fact the density of these joints is higher in certain areas of the city, as various maps in the sections to follow will depict.
Hence placing book stores close to these coffe shops as well as schools will naturally give the owners a massive advantage, as both their target demographic are more likely be in the vicinity of these locations and hence the convinience associated with the processes naturally increases.

# Methodology

1. Firstly the all the locations of the various a. Coffee shops b. Hotels c. Schools are obtained from the foursquare api and saved into their respective data frames.
2. Cleaning of these data frames is followed by the making of maps to get an idea of the various location.
3. The kmeans statistical analysis is performed to create clusters which consist of hotels,coffee shops, and schools.
4. The final visualisations of these clusters will help us to scout potential locations for new coffee shops amd book shops alike, as placing either of those in areas with morr of each other will help everyone to conduct their businesses.

So as shown by the maps density of coffee shops is relatively higher towards the Southern Part of Bombay.
Areas like Colaba, Juhu and Churchgate have an abundance of coffee places.
Adding to that there are relatively lesser number of book stores in these areas.

Additionally as most of the schools and colleges are located in the northers suburbs, opening book shops catering explicitly to the academic needs of school and college going students would make a lot more sense than opening general book stores.

The general bookstores will have a higher chance of doing well down towards the south where the coffee shops and rich people are.
The diversity in the financial demographic in the city is also well estimated by the sheer abundance of premium coffe shops down towards south. 

So as a someone looking to open bookstores: 
1. Vendors looking tos sell academic books whould look for places around the northers suburbs.
2. Bookstores which aim to sell mainly novels and those kinf of books should look for places down south.
